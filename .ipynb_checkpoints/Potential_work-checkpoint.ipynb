```python
import pandas as pd
import matplotlib.pyplot as plt
import tabula
import math
from scipy.optimize import least_squares
```

```python
#importing SST data
file = './resources/sstdata.pdf'
table = tabula.read_pdf(file,pages=73)
```

```python
#setting that data up as a df
tdf = pd.DataFrame(table[0])
tdf = tdf[["Dataset Average 1997-2016","Avg"]][0:11]
```

```python
#Sea Areas in km^2 of these 10 areas
# data taken manually from below link 
# https://marine.gov.scot/data/facts-and-figures-about-scotlands-sea-area-coastline-length-sea-area-sq-kms
area = [4279, 12044, 3473.34, 3473.33, 3473.33, 20848, 9256, 6154.5, 6154.5, 5876, 3153]
tdf["Area"] = area
```

```python
#Separating the seas into West/North/East coasts
Wcoast = tdf[0:6]
NEcoast = tdf[6:10]
SEcoast = tdf[10:11]
```

```python
#Habitat Suitibility Index function
#ins: 
#   sst - sea surface temp (in Celcius)
#   b,c - Estimated parameters 
#out: float from [0.0 - 1.0] describing habitability of the area given its sst. 
def HSI(sst,b,c):
    hsi = math.exp( (-sst**2 + 2*sst*b - b**2) / (c**2) )
    return(hsi)
```

```python
#Calculating Weights for each Sea Area by sq km

#total surface area of W(est)coast and N(orth)coast
Wtotal = sum(Wcoast["Area"])
NEtotal = sum(NEcoast["Area"])
SEWeight = 1 #There's only one area in the Ecoast so there's no need to weight that part

#actual calculation of weights
Wcoast["Weight"] = Wcoast["Area"]/Wtotal
NEcoast["Weight"] = NEcoast["Area"]/NEtotal

#Caculating the weighted avg SST 
Wcoast["wAvg"] = Wcoast["Weight"] * pd.to_numeric(Wcoast["Avg"])
NEcoast["wAvg"] = NEcoast["Weight"] * pd.to_numeric(NEcoast["Avg"])
```

```python
#The Yearly Average SST on the West Coast of Scotland (2016)
WSST = sum(Wcoast["wAvg"])
#The Yearly Average SST on the NE Coast of Scotland (2016)
NESST = sum(NEcoast["wAvg"])
#The Yearly Average SST on the SE Coast of Scotland (2016)
ESST = float(SEcoast["Avg"])
print(f'2016 NorthEast Coast Avg SST: {NESST}')
print(f'2016 West Coast Avg SST: {WSST}')
print(f'2016 SouthEast Coast Avg SST: {ESST}')
```

```python
# Potential Estimates for the parameters of the HSI
# b = 9.5 , 10.25 , 10.5 , 11 , 11.5
# c = 2.5, 2.75, 3, 3.25

# We will treat b = 10.25 as the ideal temperature of Mackerels and Herrings  
b = 10.25
# c = 2.5 streches the HSI curve to an appropriate range of habitable temperatures for these fish
c = 2.5
```

```python
#Checking HSI function
WestCoastHSI = HSI(WSST,b,c)
NECoastHSI = HSI(NESST,b,c)
SECoastHSI = HSI(ESST,b,c)
print(f'W HSI: {WestCoastHSI} :: 1 - (W HSI): {1 - WestCoastHSI}')
print(f'N HSI: {NECoastHSI} :: 1 - (N HSI): {1 - NECoastHSI}')
print(f'E HSI: {SECoastHSI} :: 1 - (E HSI): {1 - SECoastHSI}')
```

```python
#Graphical solution to find which part of scottish waters we can expect the fish.
#using 1997-2016 sst data

#Model 1

#Setting initial SST values
#Estimated Year Average SST on the West Coast of Scotland (2014)
WSST = sum(Wcoast["wAvg"]) - (2*0.0072)
#Estimated Year Average SST on the NE Coast of Scotland (2014)
NESST = sum(NEcoast["wAvg"]) - (2*0.0072)
#Estimated Year Average SST on the SE Coast of Scotland (2014)
ESST = float(SEcoast["Avg"]) - (2*0.0072)

#List of population distributions per region 
NE = [.34]
W = [.33]
SE = [.33]

#for each year from 2016-2050 (36 years)
for i in range(37):
    #simulating a increase in temperature per year by 0.0072 degrees C
    #and calculate the HSI for that region
    WSST = WSST + 0.0072
    NESST = NESST + 0.0072
    ESST = ESST + 0.0072
    
    #calc HSI for the year
    WestCoastHSI = HSI(WSST,b,c)
    NECoastHSI = HSI(NESST,b,c)
    SECoastHSI = HSI((ESST),b,c)
    
    #calc the next year's population distribution
    NE_next = NECoastHSI*NE[i] + (1-WestCoastHSI)*W[i] + (1-SECoastHSI)*SE[i]
    W_next = ((1-NECoastHSI)/2)*NE[i] + WestCoastHSI*W[i] 
    SE_next = ((1-NECoastHSI)/2)*NE[i] + SECoastHSI*SE[i]
    #save it
    NE.append(NE_next)
    W.append(W_next)
    SE.append(SE_next)
#plot it    
r1 = pd.DataFrame(list(zip(NE,W,SE)), columns = ["NorthEast", "West", "SouthEast"])
r1.cumsum()
plt.figure()
plot1 = r1.plot(grid=True)
fig1 = plot1.get_figure()
fig1.savefig('./outputs/Model1Graph.png', bbox_inches='tight')
```
